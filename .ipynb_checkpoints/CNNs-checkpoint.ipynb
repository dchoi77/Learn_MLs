# 1D CNN


## tf.keras.layers.Conv1D()

1D convolution layer (e.g. temporal convolution).

This layer creates a convolution kernel that is convolved with the layer input over a single spatial (or temporal) dimension to produce a tensor of outputs.

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Dense, Conv1D, Lambda
from tensorflow.keras.models import Sequential

input_shape = (2, 10, 5)
x = tf.random.normal(input_shape)
model = Conv1D(7, 3, activation='relu', input_shape=input_shape[1:])
y = model(x)

print(y.shape, model.kernel.shape, model.bias.shape)
```

* input_shape = (batch_size, timesteps, input_dim)


Think of a curve $\mathbf{r}(t)\in\mathbb{R}^{input\_dim}$ where $t = t_1, t_2, \ldots, t_{timesteps}$. 


* Apply a kernel of size 3 to a vector of length 10. Then we get a vector of length 8. Thus y.shape\[1\] is 8. 
* Since the number of output filters is 7, model.bias.shape is (7,).
* model.kernel.shape is \[3, 5, 7\] = \[kernel_size, num_input_filters, num_output_filters\]. 

```python
input_shape = (2, 5, 10, 20)
x = tf.random.normal(input_shape)
model = Conv1D(30, 3, activation='relu', input_shape=input_shape[2:])
y = model(x)

print(y.shape, model.kernel.shape, model.bias.shape)
```

* Since input_shape is (10, 20), batch_shape is (2, 5).

### Causal padding

```python
input_shape = (2, 10, 5)
x = tf.random.normal(input_shape)

model = Conv1D(7, 3, activation='relu', input_shape=input_shape[1:])
y = model(x)
print(y.shape)

model = Conv1D(7, 3, activation='relu', padding='valid', input_shape=input_shape[1:])
y = model(x)
print(y.shape)

model = Conv1D(7, 3, activation='relu', padding='causal', input_shape=input_shape[1:])
y = model(x)
print(y.shape)

model = Conv1D(7, 3, activation='relu', padding='same', input_shape=input_shape[1:])
y = model(x)
print(y.shape)
```

Reference: https://github.com/tensorchiefs/dl_book/blob/master/chapter_02/nb_ch02_04.ipynb

* padding="causal" results in causal (dilated) convolutions.

* With a normal causal 1D convolution of length 2 you can look back one time step. A kernel of size 2 allows to look back 1 step in time. So the $k$ convolutional layers can look back $k$ time steps. If you need to look back many timesteps you need to stack a lot of causal 1D convolutions which makes it hard to fit the model. 

<img src="images/ml_12.png">

* The trick is to use dilated convolutions. 

<img src="images/ml_13.png">

* The above uses four 1D convolution layers with the following parameters:

    * kernel_size = 2
    * padding = 'casual'
    * dilation_rate = 1, 2, 4, 8 (from bottom to top)

## Example

Reference: https://github.com/tensorchiefs/dl_book/blob/master/chapter_02/nb_ch02_04.ipynb

```python
def gen_data(sample_size=1000, seq_length=128, look_ahead=10, noise_scale=0.1, random_seed=1):
    # returns X, Y
    # X.shape: (sample_size, seq_length, 1)
    # Y.shape: (sample_size, look_ahead, 1)
    np.random.seed(random_seed)
    s = seq_length + look_ahead
    d = np.zeros((sample_size, s, 1))
    theta1 = np.linspace(0, 20*np.pi, s)
    theta2 = np.linspace(0, np.pi, s)
    for i in range(sample_size):
        start = np.random.uniform(0, 2*np.pi)
        d[i,:,0] = np.sin(start + theta1) * np.sin(start + theta2) + np.random.normal(0, noise_scale, s)
    return d[:, :seq_length], d[:, seq_length:s]
```

```python
X, Y = gen_data()
seq_length, look_ahead = X.shape[1], Y.shape[1]

plt.plot(range(seq_length), X[0,:,0], 'b-')
plt.plot(range(seq_length, seq_length + look_ahead), Y[0,:,0], 'b-', color='orange')
```

```python
ks = 5    # kernel_size

model1 = Sequential([
    # input_shape[0] = seq_length = None
    Conv1D(filters=32, kernel_size=ks, padding='causal', input_shape=(None, 1)), # (batch_size, seq_length, 32)     
    Conv1D(filters=32, kernel_size=ks, padding='causal'),                        # (batch_size, seq_length, 32)     
    Conv1D(filters=32, kernel_size=ks, padding='causal'),                        # (batch_size, seq_length, 32)
    Conv1D(filters=32, kernel_size=ks, padding='causal'),                        # (batch_size, seq_length, 32)
    Dense(1),                                                                    # (batch_size, seq_length, 1)
    Lambda(lambda x, slice_length: x[:,-slice_length:], arguments={'slice_length': look_ahead}) # (batch_size, look_ahead, 1)
]) 

model1.compile(optimizer='adam', loss='mean_squared_error')

history1 = model1.fit(X[:800], Y[:800], epochs=100, batch_size=128, validation_data=(X[800:], Y[800:]), verbose=0)
```

```python
# Using dilated convolutions

model2 = Sequential([
    # input_shape[0] = seq_length = None
    Conv1D(filters=32, kernel_size=ks, padding='causal', input_shape=(None, 1), dilation_rate=1),
    Conv1D(filters=32, kernel_size=ks, padding='causal', dilation_rate=2), 
    Conv1D(filters=32, kernel_size=ks, padding='causal', dilation_rate=4),
    Conv1D(filters=32, kernel_size=ks, padding='causal', dilation_rate=8),
    Dense(1),
    Lambda(lambda x, slice_length: x[:,-slice_length:], arguments={'slice_length': look_ahead})
]) 

model2.compile(optimizer='adam', loss='mean_squared_error')

history2 = model2.fit(X[:800], Y[:800], epochs=100, batch_size=128, validation_data=(X[800:], Y[800:]), verbose=0)
```

```python
# Prediction one after another
def predict_sequence(input_sequence, model, pred_steps):
    history_sequence = input_sequence.copy()
    pred_sequence = np.zeros(pred_steps * look_ahead)
    
    for i in range(pred_steps):
        last_step_pred = model.predict(history_sequence)[0,-look_ahead:,0]
        pred_sequence[(i * look_ahead) : ((i+1) * look_ahead)] = last_step_pred
        history_sequence = np.concatenate([history_sequence, last_step_pred.reshape(-1, look_ahead, 1)], axis=1)

    return pred_sequence
```

```python
X_test, Y_test = gen_data(sample_size=1, noise_scale=0.0, random_seed=10)

pred_steps = 12

preds1 = predict_sequence(X_test, model1, pred_steps)
preds2 = predict_sequence(X_test, model2, pred_steps)

x_test = X_test.ravel()
s = seq_length + pred_steps*look_ahead

plt.figure(figsize=(10,5))
plt.plot(range(seq_length), x_test)
plt.plot(range(seq_length, s), preds1, color='orange')
plt.plot(range(seq_length, s), preds2, color='green')
plt.ylim((-1,1))
```

* model2 can look back over longer times than model1 does.
